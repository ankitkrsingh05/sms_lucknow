```python
# 5. Write a program to flatten a nested list using recursion.
```

```python
def flatten(lst):
    res=[]
    for i in lst:
        res.extend(flatten(i)) if isinstance(i,list) else res.append(i)
    return res
flatten([1,[2,[3,4],5]])
```

```python
# 11. Write a generator that yields prime numbers up to n.
```

```python
def primes(n):
    for x in range(2,n+1):
        if all(x%i!=0 for i in range(2,int(x**0.5)+1)):
            yield x
list(primes(30))
```

```python
# 13. Implement binary search using a function.
```

```python
def binary_search(arr,x):
    l,r=0,len(arr)-1
    while l<=r:
        m=(l+r)//2
        if arr[m]==x: return m
        elif arr[m]<x: l=m+1
        else: r=m-1
    return -1
binary_search([1,3,5,7,9],7)
```

```python
# 17. Write a program to check if a string has balanced parentheses.
```

```python
s='(())()'
stack=0
balanced=True
for ch in s:
    if ch=='(': stack+=1
    else: stack-=1
    if stack<0: balanced=False
balanced and stack==0
```
