```python
def cointoss():
    return random.randint(0,1)

tosses = []
for i in range(100000):
    tosses.append(cointoss())

```

```python
sns.histplot(tosses,bins= 2)
```

```python
def dicetoss():
    return random.randint(1,6)

tosses = []
for i in range(10000000):
    tosses.append(dicetoss())
```

```python
tosses
```

```python
sns.distplot(tosses,bins= 6)
```
