```python
# 21. Recursive power function
```

```python
def power(a,b):
    return 1 if b==0 else a*power(a,b-1)
power(2,5)
```

```python
# 22. Prime numbers up to n
```

```python
def primes(n):
    return [x for x in range(2,n+1) if all(x%i!=0 for i in range(2,int(x**0.5)+1))]
primes(30)
```

```python
# 23. Recursive flatten of nested list
```

```python
def flatten(lst):
    res=[]
    for i in lst:
        res.extend(flatten(i)) if isinstance(i,list) else res.append(i)
    return res
flatten([1,[2,[3,4],5]])
```

```python
# 25. Fibonacci generator
```

```python
def fib(n):
    a,b = 0,1
    for _ in range(n):
        yield a
        a,b = b,a+b
list(fib(7))
```

```python
# 26. Generator for even numbers from list
```

```python
def evens(lst):
    for i in lst:
        if i%2==0:
            yield i
list(evens([1,2,3,4,5,6]))
```

```python
# 27. Custom iterator for reverse traversal
```

```python
class Rev:
    def __init__(self,lst): self.lst=lst; self.i=len(lst)
    def __iter__(self): return self
    def __next__(self):
        if self.i==0: raise StopIteration
        self.i-=1
        return self.lst[self.i]
list(Rev([1,2,3,4]))
```

```python
# 31. String compression using counts
```

```python
s='aaabbc'
res=''
i=0
while i<len(s):
    count=1
    while i+1<len(s) and s[i]==s[i+1]: count+=1; i+=1
    res+=s[i]+str(count)
    i+=1
res
```

```python
lst=[3,4,-7,1,2,-6]
res=[]
for i in range(len(lst)):
    s=0
    for j in range(i,len(lst)):
        s+=lst[j]
        if s==0: res.append(lst[i:j+1])
res
```
